### Algorithm

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
                
    def balanced_insert(data):
        if not data:
            return None
        data.sort()
        
        
        def build_balanced_tree(data):
            if not data:
                return None
            
            mid_index = len(data) // 2
            mid_value = data[mid_index]
            
            node = TreeNode(mid_value)
            
            left_sublist = data[:mid_index]
            right_sublist = data[mid_index + 1:]
            
            node.left = build_balanced_tree(left_sublist)
            node.right = build_balanced_tree(right_sublist)
            
            return node
        # Build the balanced tree starting from the sorted values
        return build_balanced_tree(data)
       
    def preorder_traversal(self):
        print(self.value)
        if self.left:
            self.left.preorder_traversal()
        if self.right:
            self.right.preorder_traversal()
    
    def inorder_traversal(self):
        if self.left:
            self.left.inorder_traversal()
        print(self.value)
        if self.right:
            self.right.inorder_traversal()
            
    def postorder_traversal(self):
        if self.left:
            self.left.postorder_traversal()
        if self.right:
            self.right.postorder_traversal()
        print(self.value)
    
    def find(self,value):
        if value < self.value:
            if self.left is None:
                return None
            else:
                return self.left.find(value)
        elif value > self.value:
            if self.right is None:
                return None
            else:
                return self.right.find(value)
        else:
            return self
          
    def deleteNode(self,value):
        if not self:
            return None
        
        if value < self.value:
            if self.left is not None:
                self.left = self.left.deleteNode(value)
        elif value > self.value:
            if self.right is not None:
                self.right = self.right.deleteNode(value)
        else:
            # Node to be deleted found
            if self.left is None:
                return self.right
            elif self.right is None:
                return self.left
            # if both
            current = self.left
            while current.right: current = current.right
            self.value = current.value
            self.left = self.left.deleteNode(self.value)
        return self
            
def delete_node(root, value):
    if root is not None:
        return root.deleteNode(value)
    return None

def print_tree(node, level=0, label='.'):
    prefix = ' ' * (level * 2) + label + ': '
    print(prefix + str(node.value))
    if node.left: print_tree(node.left, level + 1, 'L')
    if node.right: print_tree(node.right, level + 1, 'R')
```
